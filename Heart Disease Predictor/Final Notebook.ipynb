```python
# Impute missing values for numerical columns using median
# Median is chosen over mean because it's more robust to outliers
for col in df.select_dtypes(include=np.number).columns:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"Filled missing values in '{col}' with median: {median_val}")

# For categorical columns, use mode
for col in df.select_dtypes(include='object').columns:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
        print(f"Filled missing values in '{col}' with mode: {mode_val}")
```

```python
# Function to cap outliers using IQR
def cap_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    data[column] = np.where(data[column] < lower_limit, lower_limit,
                    np.where(data[column] > upper_limit, upper_limit, data[column]))

    print(f" Outliers in '{column}' capped using IQR method.")

for col in df.select_dtypes(include=['float64', 'int64']).columns:
    cap_outliers_iqr(df, col)
```

```python
from sklearn.preprocessing import StandardScaler

# 1. Identify numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
# 2. Initialize scaler
scaler = StandardScaler()
# 3. Fit and transform the numerical data
df[num_cols] = scaler.fit_transform(df[num_cols])
print("Standardization complete.")
print(f" Normalized columns:\n{list(num_cols)}")
```

```python
from sklearn.preprocessing import LabelEncoder

# 1. Separate binary and multi-class categorical columns
binary_cols = [col for col in df.select_dtypes(include='category') if df[col].nunique() == 2]
multi_class_cols = [col for col in df.select_dtypes(include='category') if df[col].nunique() > 2]

# 2. Label Encode binary columns
le = LabelEncoder()
for col in binary_cols:
    df[col] = le.fit_transform(df[col])
    print(f"Label Encoded: {col}")

# 3. One-Hot Encode multi-class categorical columns
df = pd.get_dummies(df, columns=multi_class_cols, drop_first=True)
print(f"\n Encoding complete. New shape: {df.shape}")
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# 1. Start: Separate features and target
X = df.drop('HeartDisease', axis=1)
y = df['HeartDisease']

# 2. Split before balancing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 3. Now balance only the training set
df_train = pd.concat([X_train, y_train], axis=1)

# Separate positive and negative
df_positive = df_train[df_train['HeartDisease'] == 1]
df_negative = df_train[df_train['HeartDisease'] == 0]

# Downsample negative class (keep 10% extra negatives)
negative_target_count = len(df_positive)
negative_keep_count = int(negative_target_count * 1.1)  # 10% more negatives
df_negative_downsampled = df_negative.sample(n=negative_keep_count, random_state=42)

# Combine positives with selected negatives
df_balanced = pd.concat([df_positive, df_negative_downsampled])

# Separate features and target again
X_balanced = df_balanced.drop('HeartDisease', axis=1)
y_balanced = df_balanced['HeartDisease']

# Apply SMOTE to training set
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_balanced, y_balanced)

print(f"âœ… Shape after resampling (train set): {X_resampled.shape}")

# 4. Balance the test set
df_test = pd.concat([X_test, y_test], axis=1)
df_test_pos = df_test[df_test['HeartDisease'] == 1]
df_test_neg = df_test[df_test['HeartDisease'] == 0]

# Downsample negatives to match number of positives
df_test_neg_downsampled = df_test_neg.sample(n=len(df_test_pos), random_state=42)

# Combine and shuffle
df_test_balanced = pd.concat([df_test_pos, df_test_neg_downsampled])
df_test_balanced = df_test_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Final test features and labels
X_test_balanced = df_test_balanced.drop('HeartDisease', axis=1)
y_test_balanced = df_test_balanced['HeartDisease']

print(f"ðŸ§® Balanced test set counts: {y_test_balanced.value_counts().to_dict()}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import joblib
import numpy as np

# Print number of positives and negatives in balanced test set
pos_count = (y_test_balanced == 1).sum()
neg_count = (y_test_balanced == 0).sum()
print(f"ðŸ§® Balanced Test Set: {pos_count} positive cases, {neg_count} negative cases")

# Simple Logistic Regression (no special parameters)
logreg_model = LogisticRegression(
    penalty='l2',           # L2 regularization
    C=0.5,                  # Regularization strength (higher = less regularized)
    solver='liblinear',     # Suitable for small datasets
    max_iter=1000,          # Number of iterations to converge
    class_weight='balanced',# Balanced class weights
    random_state=42         # Ensures reproducibility
)

# Train the model on resampled (balanced) training data
logreg_model.fit(X_resampled, y_resampled)

# Predict on balanced test set
y_pred = logreg_model.predict(X_test_balanced)

# Evaluate the model
print(f"\nâœ… Logistic Regression Evaluation:")
print(confusion_matrix(y_test_balanced, y_pred))
print(classification_report(y_test_balanced, y_pred))

# Save the model
joblib.dump(logreg_model, 'ML_model.pkl')
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import joblib

# Print number of positives and negatives in balanced test set
pos_count = (y_test_balanced == 1).sum()
neg_count = (y_test_balanced == 0).sum()
print(f"ðŸ§® Balanced Test Set: {pos_count} positive cases, {neg_count} negative cases")

# Build a deep learning model (Neural Network)
model = Sequential()

# Input layer (assuming X_resampled has 30 features)
model.add(Dense(64, input_dim=X_resampled.shape[1], activation='relu'))  # First hidden layer
model.add(Dense(32, activation='relu'))  # Second hidden layer
model.add(Dense(16, activation='relu'))  # Third hidden layer
model.add(Dense(1, activation='sigmoid'))  # Output layer (binary classification)

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model on resampled (balanced) training data
model.fit(X_resampled, y_resampled, epochs=20, batch_size=32, verbose=1)

# Predict on balanced test set
y_pred_proba = model.predict(X_test_balanced)
y_pred = (y_pred_proba >= 0.5).astype(int)

# Evaluate the model
print(f"\nâœ… Deep Learning Model Evaluation:")
print(confusion_matrix(y_test_balanced, y_pred))
print(classification_report(y_test_balanced, y_pred))

# Save the model
model.save('DL.h5')
```
